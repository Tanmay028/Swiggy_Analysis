```python
df_copy = df_copy.dropna(subset = ["city"])
```

```python
df_copy["city"].isnull().sum()
```

```python
locations = pd.DataFrame(df_copy["city"].unique())
```

```python
locations.columns = ["name"]
```

```python
locations
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="app", timeout=None)
# "app" is used to establish connection with Nominatim
# Just to avoid timeout error ...timeout =  None is done
```

```python
lat = []
lon = []

for location in locations["name"]:
    location = geolocator.geocode(location)
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
locations["latitude"] = lat
locations["longitude"] = lon
```

```python
locations
```

### How to automate your task : i.e. Perform Automation

```python
locations.isnull().sum()
```

```python
def get_heatmap(cuisine):
    cuisine_df = df[df["cuisine"].str.contains(cuisine)]
    cuisine_rest_count = cuisine_df["city"].value_counts().reset_index().rename(columns = {"index":"name","city":"Count"})
    heatmap_df = cuisine_rest_count.merge(locations, on = "name",how = "left")
    heatmap_df = heatmap_df.dropna()
    print(heatmap_df.head(4))

    basemap = folium.Map()
    HeatMap(heatmap_df[['latitude', 'longitude', "Count"]]).add_to(basemap)
    return basemap
```

```python
get_heatmap("South Indian")
```
